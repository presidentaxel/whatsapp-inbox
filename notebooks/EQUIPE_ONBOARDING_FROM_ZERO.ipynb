# WhatsApp Inbox - installation (**ce fichier suffit**)

Tu as reçu ce notebook par **e-mail** ou **Slack** : tu n’as besoin d’aucun autre document pour commencer.

**À faire dans l’ordre (une fois chaque chose) :**

1. Installer **Visual Studio Code** et ouvrir **ce fichier** dedans (voir ci-dessous).  
2. Créer un **compte GitHub**
3. Attendre l’**accès au dépôt**, puis installer **Git**, **cloner**, ouvrir le **dossier du projet** dans VS Code.  
4. Installer **Python** et **Node**, puis les commandes **backend / frontend** en bas.

**Terminal** dans VS Code : `Terminal` → `New Terminal`. Après **chaque** installation logicielle (Git, Python, Node), **ferme le terminal** et rouvre-en un nouveau (`+` à droite du panneau Terminal) pour que les commandes soient reconnues.


## 1. Visual Studio Code + ouvrir ce notebook

1. Télécharge et installe : **[code.visualstudio.com/download](https://code.visualstudio.com/download)**  
2. Lance VS Code → menu **File** → **Open File…** → choisis le fichier **`.ipynb`** que tu as enregistré depuis ta pièce jointe mail (souvent dans *Téléchargements*).  
3. Installe deux **extensions** (`Ctrl+Shift+X`, puis recherche) : **Python** et **Jupyter** (toutes deux Microsoft).

Tu peux lire tout le notebook tout de suite ; les **commandes** (`git`, `npm`, etc.) ne fonctionneront qu’après les étapes d’installation et le **clone** du dépôt (section 3).

Quand tu as reçu la confirmation d’**accès au repo** et l’**URL** exacte (si ce n’était pas déjà dans le mail), passe à la section 3.


## 3. Git, clone, ouvrir le projet

**Installer Git** : [git-scm.com/downloads](https://git-scm.com/downloads) - puis **nouveau terminal** dans VS Code :

```bash
git --version
git config --global user.name "Prénom Nom"
git config --global user.email "email@entreprise.com"
```

**Cloner** (remplace par l’URL que ton lead t’a donnée) - le plus simple dans VS Code : `Ctrl+Shift+P` → **Git: Clone** → colle l’URL → choisis un dossier parent.

Ou en ligne de commande (crée un dossier au choix, puis) :

```bash
git clone <URL_DU_DEPOT.git>
cd <nom-du-dossier-cree>
```

En HTTPS, utilise **login GitHub + PAT** si on te demande un mot de passe.

**Ouvrir le projet** : **File → Open Folder** → sélectionne la **racine** du clone (dossier qui contient `backend/`, `frontend/`, `notebooks/`). Rouvre ce notebook via `notebooks/EQUIPE_ONBOARDING_FROM_ZERO.ipynb` si besoin.

*(Option sans CLI : [GitHub Desktop](https://desktop.github.com/) pour cloner - le reste du guide reste en terminal VS Code.)*

## 6. Backend, variables d’environnement, frontend, lancer

Toutes les commandes ci-dessous partent de la **racine du clone** (dossier avec `backend/` et `frontend/`).

**Backend**

```powershell
cd backend
python -m venv .venv
.\.venv\Scripts\Activate.ps1
python -m pip install --upgrade pip
pip install -r requirements.txt
copy .env.example .env
```

*(macOS / Linux : `python3 -m venv .venv`, `source .venv/bin/activate`, `cp .env.example .env`.)*

Édite **`backend/.env`** et crée **`frontend/.env`** avec les valeurs que ton lead t’a transmises (noms des variables dans `.env.example` - pas de secrets par mail en clair).

**Frontend**

```bash
cd ../frontend
npm install
```

**Lancer** - deux terminaux (venv activé pour le backend) :

```bash
# Terminal 1 - dossier backend/
uvicorn app.main:app --reload --host 0.0.0.0 --port 8000
```

```bash
# Terminal 2 - dossier frontend/
npm run dev
```

**Webhook local** (depuis `backend/`, si besoin) :

```powershell
powershell -ExecutionPolicy Bypass -File scripts/start_webhook.ps1 --force
```
