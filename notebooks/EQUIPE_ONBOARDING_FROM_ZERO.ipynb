# Prérequis - WhatsApp Inbox

**Éditeur obligatoire : [Visual Studio Code](https://code.visualstudio.com/) ou [Cursor](https://cursor.com/).** Ouvre ce fichier `.ipynb` dedans (aperçu notebook intégré) et suis les étapes **dans l’ordre**. Toutes les commandes passent dans le **terminal intégré** (`Terminal` → `New Terminal`), sauf indication contraire.


## 1. VS Code ou Cursor

Installe l’un des deux, ouvre le dossier du dépôt cloné (**File → Open Folder**), puis ouvre ce notebook.

Extensions utiles (optionnel) : Python, ESLint, Prettier - selon ce que l’équipe utilise déjà.


## 2. Windows - PowerShell (une fois)

Pour pouvoir lancer les scripts `.ps1` du projet sans erreur :

```powershell
Set-ExecutionPolicy -Scope CurrentUser -ExecutionPolicy RemoteSigned
```

Si la politique est imposée par l’entreprise : ticket **IT** pour autoriser l’exécution côté utilisateur, ou lance les scripts avec :

```powershell
powershell -ExecutionPolicy Bypass -File .\chemin\vers\script.ps1
```


## 3. Git

- Installateur : https://git-scm.com/downloads  
- Puis dans le terminal (adapte nom / email) :

```bash
git config --global user.name "Prénom Nom"
git config --global user.email "email@entreprise.com"
```


## 4. Compte GitHub

- Création : https://github.com/signup  
- 2FA recommandée / souvent obligatoire en org : https://docs.github.com/en/authentication/securing-your-account-with-two-factor-authentication-2fa  
- Clone en HTTPS avec 2FA : **Personal Access Token** (pas le mot de passe du compte) : https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/managing-your-personal-access-tokens


## 5. Clé SSH (recommandé) + accès au dépôt

**À faire par le développeur**

1. Générer une clé : https://docs.github.com/en/authentication/connecting-to-github-with-ssh/generating-a-new-ssh-key-and-adding-it-to-the-ssh-agent  
2. Ajouter la clé **publique** sur GitHub : *Settings → SSH and GPG keys*.  
3. Tester : `ssh -T git@github.com`

**À faire par un mainteneur / admin** (tu n’as pas les droits toi-même)

- T’ajouter **collaborateur** sur le dépôt privé, ou t’ajouter à l’**équipe** GitHub qui a accès.  
- Te transmettre les **secrets** pour les fichiers `.env` (ou accès coffre d’équipe).


## 6. Python 3.11 ou plus

- Téléchargement : https://www.python.org/downloads/  
- Sous Windows : cocher **Add python.exe to PATH**.  
- Redémarrer le terminal après installation.


## 7. Node.js 18 ou plus (LTS)

- Téléchargement : https://nodejs.org/  
- Redémarrer le terminal après installation.


## 8. Docker Desktop

- https://www.docker.com/products/docker-desktop/  
- Windows : suivre l’installateur (souvent **WSL 2** : https://learn.microsoft.com/windows/wsl/install ).


## 9. Cloner le dépôt

Remplace `ORG` et `REPO` :

```bash
git clone git@github.com:ORG/REPO.git
```

ou en HTTPS :

```bash
git clone https://github.com/ORG/REPO.git
```

Ouvre **dans VS Code / Cursor** le dossier racine du clone (`File → Open Folder`).


## 10. Backend - environnement virtuel + dépendances

Dans le terminal, **depuis la racine du clone** :

**Windows (PowerShell)**

```powershell
cd backend
python -m venv .venv
.\.venv\Scripts\Activate.ps1
python -m pip install --upgrade pip
pip install -r requirements.txt
```

**macOS / Linux**

```bash
cd backend
python3 -m venv .venv
source .venv/bin/activate
python -m pip install --upgrade pip
pip install -r requirements.txt
```


## 11. Fichiers `.env`

**Windows**

```powershell
copy .env.example .env
```

**macOS / Linux**

```bash
cp .env.example .env
```

*(commande lancée dans le dossier `backend/`.)*

Remplis `backend/.env` et crée `frontend/.env` selon le **README** à la racine du projet. Sans les bonnes valeurs, l’app ne pourra pas parler à Supabase / Meta.


## 12. Frontend - dépendances npm

Depuis la racine du clone :

```bash
cd frontend
npm install
```


## 13. (Optionnel) Outils npm à la racine du repo

Uniquement si tu utilises la CLI Supabase en local (`npx supabase`, etc.) :

```bash
cd ..
npm install
```

*(remonte d’un niveau si tu étais encore dans `frontend/`.)*


## 14. Lancer le projet au quotidien

**Terminal 1 - backend** (avec le venv activé, dossier `backend/`) :

```bash
uvicorn app.main:app --reload --host 0.0.0.0 --port 8000
```

**Terminal 2 - frontend** (dossier `frontend/`) :

```bash
npm run dev
```

**Webhook / ngrok** (PowerShell, dossier `backend/`) :

```powershell
powershell -ExecutionPolicy Bypass -File scripts/start_webhook.ps1 --force
```

Détails et schéma SQL : **README** à la racine du dépôt.
